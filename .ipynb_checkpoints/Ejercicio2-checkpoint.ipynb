<h2> SVD y PCA </h2>

El dataset que se escogio contiene informacion de los casos confirmados de COVID en diferentes paises del mundo.

```python
import pandas as pd

data = "CONVENIENT_global_confirmed_cases.csv"

df = pd.read_csv(data)
df = df.dropna()
df
```

Se definen las 'features' (las columnas del dataframe data a las que les vamos a aplicar el PCA

```python
features = ["Afghanistan","Australia","Argentina","Mexico","Venezuela","Colombia","US","Uruguay"]

x = df.loc[:, features]

y = df.loc[:, ['Mexico']].values
x
```

<h3>Â¿Que efecto hace la funcion StandarScaler a x? Â¿PodrÃ­as explicarlo en tÃ©rminos de lo
visto en Ã¡lgebra lineal y estadÃ­stico?</h3>

La funcion StandarScaler lo que hace es estandarizar el conjunto de datos, esto implica transformar los datos para que la media de los valores observados sea 0 y la desviacion estandar sea 1. La razon porque la estandarizacion es necesaria antes de realizar un PCA es que el analisis de componentes principales es muy sensible a las variaciones.

```python
from sklearn.preprocessing import StandardScaler

x = StandardScaler().fit_transform(x)

x
```

<h3>Â¿CuÃ¡l es el mÃ¡ximo de componentes principales que puedo pedir?</h3>

Por lo general dada una matriz de datos de dimensiones n x p, el numero de ocmponentes principales que se pueden calcular es como mÃ¡ximo de n-1 o p (el menor de los dos valores es el limite). En este caso p es mayor por lo que el numero maximo a tomar seria de n. Que en este caso es 8.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=4)
principalComponents = pca.fit_transform(x)
principalComponents
```

```python
matrixComponents = pd.DataFrame(data = principalComponents, columns = ["CP1", "CP2","CP3","CP4"])
matrixComponents
```

```python
print("La suma de las varianzas de las 4 componentes principales: \n", sum(pca.explained_variance_ratio_))
```

<h3>Â¿CÃ³mo sÃ© quÃ© tantas componentes es pertinente tomar en cuenta para mis datos?</h3>

Hay que saber la varianza explicada de cada componente, en este caso con las primeras 4 ya explica el 90% de la varianza.

```python
import matplotlib.pyplot as plt

plt.plot(range(0,4), pca.explained_variance_ratio_)
plt.ylabel('Explained Variance')
plt.xlabel('Principal Components')
plt.show()
```

En esta grafica podemos ver que la primer componente explica la mayor varianza y va decreciendo.
