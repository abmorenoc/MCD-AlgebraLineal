<h2> Ejercicio 1: Calculando SVD de la matriz [[1,2], [2,3]] </h2>

En este ejercicio se pretende calcular el SVD de una matriz y ver que efecto tiene sobre un conjunto de vectores en $R^2$ 
(Circulo con centro en $[0,0]$ y radio = 1)

```python
import numpy as np
from matplotlib.pyplot import *
import math as mt
```

Primero calculamos las matrices $ U S V^T$ 

```python
A = np.matrix([[1, 2], [2, 3]])
u, s, v = np.linalg.svd(A)

print("u: ", '\n',u)
print("s: ", '\n', np.diag(s))
print("v: ", '\n', v.T)

```

Calculamos la matriz de puntos que froman el ciculo con centro en [0,0] y Radio = 1

```python
ndiv=100
r=1.0
theta=np.linspace(0,2*np.pi,ndiv)
x=r*np.cos(theta)
y=r*np.sin(theta)    

xy = np.vstack((x, y)).T

fig = figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x,y,'b.')
axis('equal')
```

Graficamos el efecto sobre el conjunto de los vectores en $R^2$ al aplicarle $V^T$ luego $S$ y luego $U$. La nueva matriz resultante se llamara $B$.

```python

B = np.dot(u, np.dot(np.diag(s), np.dot(v, xy.T))).T

fig = figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x,y,'b.')
axis('equal')
ax.plot(B[:,0], B[:,1], 'g-')
axis('equal')
grid('on')
```

Ahora aplicaremos directamente la matriz $A$

```python
A2 = np.dot(xy, A)
fig = figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x,y,'b.')
axis('equal')

ax.plot(A2[:,0], A2[:,1], 'r-')
axis('equal')
grid('on')
```

Como se puede observar realizaron la misma transformacion en el conjunto de puntos. Por lo tanto la matriz inicial $A$ y la
matriz resultante de $U S V^T$ son iguales.

<h2> Calculando la matriz de <b>Rank 1</b> que mejor se aproxime a A </h2>


```python
Arank1 = u.A[0] * v.A[0] * s[0]

print("Matriz de rango 1 que se aproxima a A: \n",Arank1)
```
